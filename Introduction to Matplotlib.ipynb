## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]
# 2. setup plot
fig, ax = plt.subplots(figsize = (5,5))#(width,height)
#3. plot data
ax.plot(x,y)
#4. customize plot
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")
#5. save and show (you the whole figure)
fig.savefig("images/sample-plot.png")
```

## Two option subplot

```python
# subplot option 1
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize =(10,5) )
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(Nut_Butter.keys(),Nut_Butter.values());
ax4.hist(np.random.randn(1000));
```

```python
# 2nd method
fig,ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))
ax[0][0].plot(x,x/2);
ax[0][1].scatter(np.random.random(10),np.random.random(10));
ax[1][0].bar(Nut_Butter.keys(),Nut_Butter.values());
ax[1][1].hist(np.random.randn(1000));
```

## Which one to use (pyplot vs matplotlib oo method??)
* when ploting something quick use okay to use pyplot method
* when plotting something more advanced use oo method

```python
#pyplot
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.plot(kind = "scatter",
              x = "age",
               y = "chol",
               c = "target");
```

```python
# oo method
fig,ax  =plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
             x= "age",
             y = "chol",
             c ="target",
             ax =ax);
#ax.set_xlim([45,100]);
```

```python

# oo method
fig,ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],)
# customize the plot
ax.set(title = "Heart disease and cholestrol level",
       xlabel = "age",
       ylabel = "cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(),title ="target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = ':');
```

```python
# oo method
fig,(ax1,ax2) =plt.subplots( nrows = 2,
                             ncols = 1,
                         figsize = (10,10),
                           sharex = True)
# enter data
scatter = ax1.scatter(x= over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])
# customize the plot
ax1.set(title = "heart disease and cholestrol level",
         ylabel = "cholestrol");
# add a legend
ax1.legend(*scatter.legend_elements(),title = 'target');
# add a horizontal line
ax1.axhline(over_50["chol"].mean(),
             linestyle = ":");

#enter the data for 2nd graph
scatter = ax2.scatter(x= over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])
#customize the plot
ax2.set( title = "heart disease and max heart level",
         xlabel = "age",
         ylabel = "thalach");

# add a legend
ax2.legend(*scatter.legend_elements(),title = 'target');
# add a horizontal line
ax2.axhline(over_50["thalach"].mean(),
            linestyle = ":");

# add a heading
fig.suptitle("Heart disease analysis",fontsize = 16,fontweight = "bold");
```

```python
plt.style.use('seaborn-whitegrid')
# oo method
fig,ax = plt.subplots(figsize = (10,6))

scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = 'winter')#changes color of black and white dot
# customize the plot
ax.set(title = "Heart disease and cholestrol level",
       xlabel = "age",
       ylabel = "cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(),title ="target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = ':');
```

```python
# oo method
fig,(ax1,ax2) =plt.subplots( nrows = 2,
                             ncols = 1,
                         figsize = (10,10),
                           sharex = True)
# enter data
scatter = ax1.scatter(x= over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                      cmap = "winter")#changes colors of dot
# customize the plot
ax1.set(title = "heart disease and cholestrol level",
         ylabel = "cholestrol");
# Changes x limits
ax1.set_xlim([50,80])
# add a legend
ax1.legend(*scatter.legend_elements(),title = 'target');
# add a horizontal line
ax1.axhline(over_50["chol"].mean(),
             linestyle = ":");

#enter the data for 2nd graph
scatter = ax2.scatter(x= over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = "winter")#changes colors of dot
#customize the plot
ax2.set( title = "heart disease and max heart level",
         xlabel = "age",
         ylabel = "thalach");
# Changes x limits
ax2.set_xlim([50,80])
#changes y limits
ax2.set_ylim([60,200])

# add a legend
ax2.legend(*scatter.legend_elements(),title = 'target');
# add a horizontal line
ax2.axhline(over_50["thalach"].mean(),
            linestyle = ":");

# add a heading
fig.suptitle("Heart disease analysis",fontsize = 16,fontweight = "bold");
```
